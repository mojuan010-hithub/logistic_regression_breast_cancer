# Logistic regression for breast cancer

## 1. import libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

## 2. read and overview dataset

```python
url = "https://raw.githubusercontent.com/mojuan010-hithub/logistic_regression_breast_cancer/refs/heads/main/breast_cancer.csv"
df = pd.read_csv(url)
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df["Diagnosis"].value_counts()
```

## 3. preprocess features and target

```python
x = df.drop(columns = ["Diagnosis","ID"])
x.head(5)
```

```python
y = df["Diagnosis"]
y.head(5)
```

```python
# switch catergorial target to binary
y = np.where(y=="M",1,0)
y
```

```python
# standardize the features
stan_scal = StandardScaler()
x = stan_scal.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print("x_train size:",x_train.shape)
print("x_test size:",x_test.shape)
print("y_train size:",y_train.shape)
print("y_test size:",y_test.shape)
```

## 4. linear regression predict

```python
log_reg = LogisticRegression()
log_reg
```

```python
log_reg.fit(x_train,y_train)
```

```python
y_train_pred = log_reg.predict(x_train)
y_train_pred
```

```python
y_test_pred = log_reg.predict(x_test)
y_test_pred
```

## 5. validation using accuracy_score and confusion matrix

5.1 confusion matrix

```python
y_train_confusion_matrix = confusion_matrix(y_train,y_train_pred)
y_train_confusion_matrix
```

```python
y_test_confusion_matrix = confusion_matrix(y_test,y_test_pred)
y_test_confusion_matrix
```

5.2  average accuracy_score

```python
y_train_accuracy_score_lst=[]
y_test_accuracy_score_lst=[]


for i in range(0,10):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
  log_reg = LogisticRegression()
  log_reg.fit(x_train,y_train)

  y_train_pred = log_reg.predict(x_train)
  y_train_accuracy_score = accuracy_score(y_train,y_train_pred)
  y_train_accuracy_score_lst.append(y_train_accuracy_score)

  y_test_pred = log_reg.predict(x_test)
  y_test_accuracy_score = accuracy_score(y_test,y_test_pred)
  y_test_accuracy_score_lst.append(y_test_accuracy_score)

# print("train accuracy score list:",y_train_accuracy_score_lst)
# print("test accuracy score list:",y_test_accuracy_score_lst)

y_train_accuracy_score_avg = np.mean(y_train_accuracy_score_lst)
y_test_accuracy_score_avg = np.mean(y_test_accuracy_score_lst)

print("average training data accuracy score:",y_train_accuracy_score_avg)
print("average testing data accuracy score:",y_test_accuracy_score_avg)

```
